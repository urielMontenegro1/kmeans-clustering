```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import adjusted_rand_score





opc=6


#seleccionamos  como menu la opcion que queremos de los siguientes casos
if opc == 1:
  # Extraemos vectores
  df = pd.read_csv('df_agrupacion_1.cvs')
  x=np.asanyarray(df.drop(columns=['y']))
  y=np.asanyarray(df[['y']])
elif opc==2:
  # Extraemos vectores
  df = pd.read_csv('df_agrupacion_2.cvs')
  x=np.asanyarray(df.drop(columns=['y']))
  y=np.asanyarray(df[['y']])
elif opc==3:
  # Extraemos vectores
  df = pd.read_csv('df_agrupacion_3.cvs')
  x=np.asanyarray(df.drop(columns=['y']))
  y=np.asanyarray(df[['y']])
elif opc==4:
  # Extraemos vectores
  df = pd.read_csv('df_agrupacion_4.cvs')
  x=np.asanyarray(df.drop(columns=['y']))
  y=np.asanyarray(df[['y']])
elif opc==5:
  # Extraemos vectores
  df = pd.read_csv('df_agrupacion_5.cvs')
  x=np.asanyarray(df.drop(columns=['y']))
  y=np.asanyarray(df[['y']])  

else:
  # Cambiamos el grado del polinomio del cuarto ejercicio
  # Extraemos vectores
  df = pd.read_csv('df_agrupacion_6.cvs')
  x=np.asanyarray(df.drop(columns=['y']))
  y=np.asanyarray(df[['y']])


print('n = '+str(np.max(y)+1))


plt.figure()
plt.grid()
plt.xlabel('x1')
plt.ylabel('x2')

plt.plot(x[:,0],x[:,1],'bo')
plt.show()
```

```python
def Grafica(model,n,x,y,title):
  plt.figure()
  plt.grid()
  plt.xlabel('x1')
  plt.ylabel('x2')
  plt.title(title)

  for i in range(n):
    plt.plot(x[y==i,0],x[y==i,1],'o')

    plt.plot(x[y==1,0],x[y==1,1],'rx')

    plt.show()
```

```python
from sklearn.cluster import KMeans

n=3
model=KMeans(n_clusters=n)
model.fit(x)

y_pred=model.predict(x)

score=adjusted_rand_score(y.ravel(),y_pred)
print('score = ',score)

Grafica(model,n,x,y_pred,'KMeans')
```

```python
from sklearn.cluster import MeanShift

model=MeanShift()
model.fit(x)

y_pred = model.predict(x)
n=np.max(y_pred)+1

score=adjusted_rand_score(y.ravel(),y_pred)
print('score = ',score)


Grafica(model,n,x,y_pred,'Mean Shift')
```

```python
from sklearn.cluster import SpectralClustering

n=2;

model=SpectralClustering(n_clusters=n)

model.fit(x)

y_pred = model.labels_


score=adjusted_rand_score(y.ravel(),y_pred)
print('score = ',score)


Grafica(model,n,x,y_pred,'Spectral Clustering')
```

```python
from sklearn.cluster import AgglomerativeClustering

n=2;

model=AgglomerativeClustering(n_clusters=n)

model.fit(x)

y_pred = model.labels_


score=adjusted_rand_score(y.ravel(),y_pred)
print('score = ',score)


Grafica(model,n,x,y_pred,'Agglomerative Clustering')
```

```python
from sklearn.cluster import DBSCAN

model=DBSCAN(eps=0.2)
model.fit(x)

y_pred = model.labels_
n=np.max(model.labels_)+1

score=adjusted_rand_score(y.ravel(),y_pred)
print('score = ',score)


Grafica(model,n,x,y_pred,'DBSCAN')
```

```python
from sklearn.cluster import OPTICS

model=OPTICS(min_cluster_size=0.1)
model.fit(x)

y_pred = model.labels_
n=np.max(model.labels_)+1

score=adjusted_rand_score(y.ravel(),y_pred)
print('score = ',score)


Grafica(model,n,x,y_pred,'OPTICS')
```

```python
from sklearn.mixture import GaussianMixture

n=2

model=GaussianMixture(n_components=n)
model.fit(x)

y_pred = model.predict(x)


score=adjusted_rand_score(y.ravel(),y_pred)
print('score = ',score)


Grafica(model,n,x,y_pred,'Gaussian Mixture')
```

```python
from sklearn.cluster import Birch

n=2

model=Birch(n_clusters=n)
model.fit(x)

y_pred = model.predict(x)


score=adjusted_rand_score(y.ravel(),y_pred)
print('score = ',score)


Grafica(model,n,x,y_pred,'Birch')
```
